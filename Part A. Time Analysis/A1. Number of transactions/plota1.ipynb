```python
import matplotlib.pyplot as plt
import calendar

data = [["09/2018", 16056742], ["06/2017", 7244657], ["12/2015", 347092], ["06/2018", 22471788], ["08/2015", 85609], ["08/2018", 19842059], ["01/2017", 1409664], ["09/2017", 10679242], ["10/2017", 12602063], ["08/2017", 10523178], ["07/2016", 1356907], ["12/2016", 1316131], ["09/2015", 173805], ["10/2018", 17056926], ["01/2018", 33504270], ["03/2017", 2426471], ["12/2018", 17107601], ["04/2016", 1023096], ["11/2018", 16713911], ["06/2016", 1351536], ["01/2016", 404816], ["03/2016", 917170], ["03/2018", 20261862], ["07/2017", 7835875], ["02/2018", 22231978], ["02/2017", 1410048], ["02/2016", 520040], ["10/2016", 1329847], ["07/2018", 19937033], ["10/2015", 205045], ["11/2015", 234733], ["12/2017", 26732085], ["11/2016", 1301586], ["09/2016", 1387412], ["11/2017", 15292269], ["05/2017", 4245516], ["04/2017", 2539966], ["01/2019", 1002431], ["05/2016", 1346796], ["08/2016", 1405743], ["04/2018", 20876642], ["05/2018", 25105717]]

# Sort the data by month and year
sorted_data = sorted(data, key=lambda x: (int(x[0].split('/')[1]), int(x[0].split('/')[0])))

# Extract the year and month from the date string
months = [x[0] for x in sorted_data]
years = [int(x.split('/')[1]) for x in months]
months = [calendar.month_name[int(x.split('/')[0])] for x in months]

# Extract the transaction counts
transactions = [x[1] for x in sorted_data]

# Create a bar plot of the transaction counts by month
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(range(len(sorted_data)), transactions)
ax.set_xticks(range(len(sorted_data)))
ax.set_xticklabels([f"{months[i]} {years[i]}" for i in range(len(sorted_data))], rotation=90)
ax.set_title('Number of Transactions by Month')
ax.set_xlabel('Month-Year')
ax.set_ylabel('Number of Transactions')
ax.grid(axis='y')

# Save plot to file
plt.savefig("plot.png", bbox_inches='tight')

plt.show()


```
